# Plot the positions and s21 of the 1st low speed area 
(original and after shift)

```python
def plot_p_s_time(pos,pos_shift,s21):
    
    """
    plot the positions and scattering parameters with respect to time
    """
    
    fig = plt.figure(figsize=(12,10))
    fig.suptitle('positions and scattering parameters with respect to time', fontsize=16)

    # original
    ax1 = plt.subplot(2,1,1)
    ax1.plot(pos,'.',label='pos')
    ax1.plot(s21,'.',label='s21')
    ax1.vlines(np.argmin(pos),-.15,.3,colors='r')
    ax1.vlines(np.argmax(pos),-.15,.3,colors='r')
    ax1.title.set_text('original')
    ax1.legend()

    # after shift
    ax2 = plt.subplot(2,1,2)

    ax2.plot(pos_shift,'.',label='pos')
    ax2.plot(s21,'.',label='s21')
    ax2.vlines(np.argmin(pos_shift),-.15,.3,colors='r')
    ax2.vlines(np.argmax(pos_shift),-.15,.3,colors='r')
    ax2.title.set_text('after shift')
    ax2.legend()
```

```python
pos = h51['pos1'][F1:L1]/1000
pos_shift = h51['pos1'][F1+SHIFT1:L1+SHIFT1]/1000
s21 = abs(h51['S21_P1_LEM1'][F1:L1,fidx])

%matplotlib inline
plot_p_s_time(pos,pos_shift,s21)
```

```python
def plot_p_s(pos,pos_shift,s21):
    
    """
    plot the scattering parameters with respect to postions at LAST
    """
    
    fig = plt.figure(figsize=(12,8))
    fig.suptitle('scattering parameters with respect to positions', fontsize=16)
    # original
    ax1 = plt.subplot(2,1,1)
    ax1.plot(pos,s21,'.',label='s21')
    ax1.title.set_text('original')
    ax1.legend
    
    # plots after shift
    ax2 = plt.subplot(2,1,2)
    ax2.plot(pos_shift,s21,'.',label='s21')
    ax2.title.set_text('after shift')
    ax2.legend
```

```python
pos = h51['pos1'][F1:L1]
s21 = abs(h51['S21_P1_LEM1'][F1:L1,fidx])
pos_shift = h51['pos1'][F1+SHIFT1:L1+SHIFT1]

%matplotlib inline
plot_p_s(pos,pos_shift,s21)
```

```python
def plot_p_s_process_accu(hdf5,first,last,interval,fidx):
    
    """
    plot the scattering parameters with respect to postions during timestamp FIRST to LAST (accumulative)
    """
    
    %matplotlib inline
    plt.figure(figsize=(20,40))
    for i in range(1,(last-first)//interval+1):
        ax=plt.subplot((last-first)//interval,4,i) 
        ax.plot(hdf5['pos1'][first:first+interval*i],abs(hdf5['S21_P1_LEM1'][first:first+interval*i,fidx]),'.',label='s21')
        ax.set_xlim(-150, +150)
        
def plot_p_s_process_interval(hdf5,first,last,interval,fidx):
    
    """
    plot the scattering parameters with respect to postions during timestamp FIRST to LAST ((only show points within interval))
    """
    
    %matplotlib inline
    plt.figure(figsize=(20,40))
    for i in range(1,(last-first)//interval+1):
        ax=plt.subplot((last-first)//interval,4,i) 
        ax.plot(hdf5['pos1'][first+interval*(i-1):first+interval*i],abs(hdf5['S21_P1_LEM1'][first+interval*(i-1):first+interval*i,fidx]),'.',label='s21')
        ax.set_xlim(-150, +150)
```

```python
plot_p_s_process_accu(h51,F1,L1,100,fidx)
```

```python
plot_p_s_process_interval(h51,F1,L1,100,fidx)
```
